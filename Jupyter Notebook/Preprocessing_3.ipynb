## Extracting features of 2018 movies from Wikipedia

```python
# there is no sufficient data on kaggle for movies after 2017 , so we are extracting the data from wikipidea
# there are 4 tables , so we just collected them into a single-single dataframe

link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
# appending all the dataframes into a single dataframe
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_id(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].id
    except:
        return np.NaN
```

```python
# based on the movie title , storing the id for movie
df['id'] = df['Title'].map(lambda x: get_movie_id(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_avg_vt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_average
    except:
        return np.NaN
```

```python
df['vote_average'] = df['Title'].map(lambda x: get_movie_avg_vt(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_vt_cnt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_count
    except:
        return np.NaN
```

```python
df['vote_count'] = df['Title'].map(lambda x: get_movie_vt_cnt(str(x)))
```

```python
# by the use of this function , we are extracting the movie related information based on their movie ID with the help of API KEY

from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
# based on the movie title , storing the geners for movie
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
df
```

```python
# these are the important columns for the 2018 movies data
df_2018 = df[['id','Title','Cast and crew','genres','vote_count','vote_average']]
```

```python
df_2018
```

```python
# extracting movie director name 
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
# extracting movie actror1 name 
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name'] = df_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
# extracting movie actror2 name 
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
# extracting movie actror3 name 
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_2018
```

```python
df_2018 = df_2018.rename(columns={'Title':'movie_title'})
```

```python
# just taking the important columns for further process
new_df18 = df_2018.loc[:,['id','director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','vote_count','vote_average']]
```

```python
new_df18
```

```python
# replacing the null values with the unknown
new_df18['actor_2_name'] = new_df18['actor_2_name'].replace(np.nan, 'unknown')
new_df18['actor_3_name'] = new_df18['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_df18['movie_title'] = new_df18['movie_title'].str.lower()
```

```python
# making a new column and Concatenate all the columns into that single column
new_df18['comb'] = new_df18['actor_1_name'] + ' ' + new_df18['actor_2_name'] + ' '+ new_df18['actor_3_name'] + ' '+ new_df18['director_name'] +' ' + new_df18['genres']
```
